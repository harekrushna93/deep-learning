```python
import numpy as np
import matplotlib.pyplot as plt
no_layer = 3
input = np.array([1,1])
m_input = np.insert(input,0,-1).reshape(1,-1)
output = 0
n = np.array([2,2,1]) # no of nurons in each layer, layer 1 = 3, layer2 = 2, layer 3 = 1
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
def grad_e(t):
    return t-t*t
```

```python
w1 = np.array([[0.8,0.5,0.4],[-0.1,0.9,1]])
w2 = np.array([0.3,-1.2,1.1])
```

```python
w1 = np.random.random((n[1],n[0]+1))
w2 = np.random.random((n[2],n[1]+1))
```

```python
e = []
l = 1
itr = 1000
for i in range(itr):
    print('in itration',i)
    print('*******************************')
    a2 = sigmoid(np.dot(w1,np.insert(input,0,-1)))
    print('output in layer:1 is ',a2)
    a3 = sigmoid(np.dot(w2,np.insert(a2,0,-1)))
    print('output in layer:2 is ',a3)
    error = output - a3
    print('EEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEE',error)
    e.append(abs(error))
    delta3 = grad_e(a3)*error
    print('delta3:::::::::',delta3)
    delta2 = ((grad_e(a2)*np.delete(w2,0))*delta3).reshape(-1,1)
    print('delta2:::::::::',delta2)
    deltaw2 = l*delta3*np.insert(a2,0,-1)
    print('deltaw2::::::::',deltaw2)
    deltaw1 = l*np.dot(delta2,m_input)
    print("deltaw1::::::::",deltaw1)
    w1 = w1 + deltaw1
    w2 = w2 + deltaw2
    print('w1=',w1)
    print('w2=',w2)
```

```python
plt.plot(np.arange(itr),e)
```
